## build gnb from scrach

### first calculate the mean and standard devivation, of each feature for each class

```python
labels = set(np_labels)
labels
```

```python
params = {}
for l in labels:
  params[l] = {
    "mean": np.mean(np_train[np_labels == l], axis=0),
    "std": np.std(np_train[np_labels == l], axis=0)
  }
params
```

### calculate the likelihood of each feature

```python
def gaussian(x, mean, var):
  return np.exp(- np.square(x - mean) / (2 * var)) / np.sqrt( 2 * np.pi * var)
```

```python
keep_mean = params["keep"]["mean"]
keep_var = params["keep"]["std"]
keep_mean, keep_var
# [x for x in np_train[:, 0]]

[gaussian(x, keep_mean[0], keep_var[0]) for x in np_train[:, 0]]


# [lambda x: ]
```
